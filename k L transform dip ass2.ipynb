```python
img1=np.double((img1).reshape(1,4096))
img2=np.double((img2).reshape(1,4096))
img3=np.double((img3).reshape(1,4096))
img4=np.double((img4).reshape(1,4096))
img5=np.double((img5).reshape(1,4096))
img6=np.double((img6).reshape(1,4096))
#----------------------------------------------------
img=np.concatenate((img1,img2),axis=0)
img=np.concatenate((img,img3),axis=0)
img=np.concatenate((img,img4),axis=0)
img=np.concatenate((img,img5),axis=0)
img=np.concatenate((img,img6),axis=0)
means=img.mean(0)
print(img)
print(means.shape)
# print(means)
# img=img.T
# print(img.shape)
# img
# img111=img[:,0]
# figure=plt.figure(figsize=(10,10))
# plt.subplot(2,3,1),plt.imshow(np.uint8(img111.reshape(64,64)),cmap='Greys_r')
```

```python
for i in range (6):
    img[i,:]=img[i,:]-means
cmat=np.dot((img.T),img)
cmat=cmat/5
evals,evec=np.linalg.eigh(cmat)
print(evals)
print(evec)
```

```python
#evals=evals[::-1]
#evec=evec[::-1]
print(evals[:10])
print(evec)
xx=np.arange(0,4096)
plt.scatter(xx,evals)
```

```python
basis=evec[:,4090:]
y=np.dot(img,basis)
res=np.dot(y,(basis.T))
for i in range (6):
    res[i,:]=res[i,:]+means
#res=np.uint8(res)
print(res)
```

```python
img11=res[0,:]
img11 =img11.reshape((64,64))
img22=res[1,:]
img22 = img22.reshape((64,64))
img33=res[2,:]
img33 = img33.reshape((64,64))
img44=res[3,:]
img44 =  img44.reshape((64,64))
img55=res[4,:]
img55 =  img55.reshape((64,64))
img66=res[5,:]
img66 =  img66.reshape((64,64))
print(img66)
```
